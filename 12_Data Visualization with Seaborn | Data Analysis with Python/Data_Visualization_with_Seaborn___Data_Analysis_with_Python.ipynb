```python
# Importing Seaborn
import seaborn as sns

# Importing Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import IPython
from IPython.display import Image
```

## Relational Plots API Interface

Relplot: Designed for visualizing different statistical relationships.

```python
# Loading Example Data
df = sns.load_dataset("tips")
df.head(2)
```

```python
sns.relplot(
    data=df,
    x='total_bill', y='tip', col='sex',
    hue='smoker', style="day"
)
```

```python
# Scatterplot
diamond=sns.load_dataset("diamonds")
diamond.head()
```

```python
sns.scatterplot(
    data=diamond,
    x='carat', y='price',
    hue='clarity'
)
```

```python
zomato=pd.read_csv('/content/drive/MyDrive/LOCTEC/DAWP 2023/Dataset2/zomato/zomato.csv')
zomato.head()
```

```python
sns.scatterplot(
    data=zomato,
    x='rate (out of 5)', y='avg cost (two people)',
    hue='online_order'
)
```

```python
# Lineplot
sns.lineplot(
    data=zomato,
    x='rate (out of 5)', y='restaurant type',
    hue='online_order'
)
```

## Distributional Plots (Distplot)

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# displot
sns.displot(
    penguins,
    x='island', col='species')
```

```python
# Histplot
sns.histplot(
    diamond, x='price', hue='cut'
)
```

```python
# Histplot
#sns.set_theme(style='ticks')
sns.histplot(
    zomato, y='area', hue='online_order'
)
```

### KDE Plot

```python
# KDE Plot
#sns.set_theme(style='ticks')
sns.displot(
    zomato, y='avg cost (two people)',
    x='num of ratings',
    bw_adjust=.90,
    multiple='stack',
    kind='kde'
)
```

```python
penguins.head()
```

## KDE Plot

Analogous to Histogram by taking the dense esimation prediction.

```python
#sns.set_theme(style='ticks')
sns.displot(
    penguins, x='flipper_length_mm',
    bw_adjust=.90,
    multiple='stack',
    kind='kde'
)
```

## ecdfplot Plot


```python
# ecdfplot Plot
sns.displot(
    penguins, x='flipper_length_mm',
    kind='ecdf'
)
```

## rugplot Plot

```python
# rugplot Plot
sns.rugplot(
    penguins, x='flipper_length_mm',
    height=0.55
)
```

# Categorical Plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
# Categorical Scatterplots
## catplot
sns.catplot(
    data=tips,
    x='day',
    y='total_bill'
)
```

```python
# catplot

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    jitter=False
)
```

```python
# Categorical scatter
## swarmplot
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='swarm'
)
```

```python
# catplot

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='swarm',
    hue='sex'
)
```

```python
# Categorical scatter
## Stripplot
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='strip'
)
```

image.png

```python
# Categorical Distribution plots
## Boxplot
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="box",
    hue="smoker"

)
```

```python
# Categorical Distribution plots
## Boxenplot
sns.boxenplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker"
)
```

```python
# Categorical Distribution plots
## Violin plot
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="violin",
    hue="smoker"
)
```

```python
# Categorical Distribution plots
## Violin plot
sns.violinplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker"
)
```

```python
# Categorical Estimation plots
## pointplot
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    kind='point'
)
```

```python
# Categorical Estimation plots
## barplot
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    kind='bar'
)
```

```python
# Categorical Estimation plots
## countplot
sns.catplot(
    data=tips,
    x="day",
    hue="smoker",
    kind='count'
)
```
